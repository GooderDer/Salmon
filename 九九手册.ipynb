下面代码输出什么？为什么？

```python
def func1(a, lst=[]):
    lst.append(a)
    return lst

def func2(a, lst=None):
    if lst is None:
        lst = []
    lst.append(a)
    return lst

print("func1: ", func1(1))
print("func1: ", func1(2))
print("func1: ", func1(3))

print("func2: ", func2(1))
print("func2: ", func2(2))
print("func2: ", func2(3))
```
